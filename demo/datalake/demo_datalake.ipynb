# Data Lake Demo Filler

```python
import boto3
import pandas as pd
import random
from faker import Faker
from io import BytesIO
from random import randint

# Konfiguration
MINIO_URL = "http://localhost:9000"
ACCESS_KEY = "admin"
SECRET_KEY = "password123"

fake = Faker('de_DE')

# S3 Client Initialisierung
s3_client = boto3.client(
    "s3",
    endpoint_url=MINIO_URL,
    aws_access_key_id=ACCESS_KEY,
    aws_secret_access_key=SECRET_KEY,
)

def upload_to_minio(df, bucket_name, file_name):
    # Bucket erstellen, falls er nicht existiert
    try:
        s3_client.head_bucket(Bucket=bucket_name)
    except:
        s3_client.create_bucket(Bucket=bucket_name)
        print(f"Bucket '{bucket_name}' erstellt.")

    # CSV im Speicher erstellen und hochladen
    csv_buffer = BytesIO()
    df.to_csv(csv_buffer, index=False)
    s3_client.put_object(Bucket=bucket_name, Key=file_name, Body=csv_buffer.getvalue())
    print(f"Datei {file_name} in {bucket_name} hochgeladen.")


def generate_customers(n=1000):
    data = [{
        "customer_id": i,
        "firtname": fake.first_name(),
        "lastname": fake.name(),
        "email": fake.email(),
        "phone": fake.phone_number(),
        "latitutde": fake.latitude(),
        "longitude": fake.longitude(),
        "address_street": fake.street_address(),
        "city": fake.city(),
        "signup_date": fake.date_this_decade(),
        "last_ip": fake.ipv4(),
        "username": fake.user_name(),
        "useragent": fake.user_agent(),
    } for i in range(10000, 10000 + n)]
    return pd.DataFrame(data)


def generate_products(n=1000):
    categories = ["Elektronik", "Software", "Hardware", "Service"]
    data = [{
        "product_id": f"PROD-{i:03}",
        "product_name": fake.catch_phrase(),
        "category": random.choice(categories),
        "length": random.uniform(100.0, 1000.0),
        "width": random.uniform(100.0, 1000.0),
        "weight": random.uniform(10.0, 100.0),
        "color": fake.hex_color(),
        "price": round(random.uniform(10.0, 500.0), 2)
    } for i in range(10000, 10000+n)]
    return pd.DataFrame(data)


def generate_order_detail(n=50000):
    data = [{
        "order_id": 5000 + i,
        "customer_id": random.randint(10000, 10999),
        "product_id": f"PROD-{random.randint(10000, 10999):03}",
        "amount": random.randint(1, 5),
        "sale_price": round(random.uniform(10.0, 500.0), 2),
        "order_date": fake.date_this_year()
    } for i in range(n)]
    return pd.DataFrame(data)


def generate_order(n=50000):
    data = [{
        "order_id": 5000 + i,
        "customer_id": random.randint(10000, 10999),
        "order_date": fake.date_this_year()
    } for i in range(n)]
    return pd.DataFrame(data)


def generate_logs(n=100000):
    data = [{
        "timestamp": fake.date_time_this_month(),
        "machine_id": f"M-{random.randint(1, 100):04}",
        "sensor": random.choice(["temp", "hum", "vib", "acc", "dec"]),
        "sensor_value": round(random.uniform(50.0, 120.0), 1),
        "status": random.choices(["OK", "WARNING", "ERROR"], weights=[0.8, 0.15, 0.05])[0]
    } for i in range(n)]
    return pd.DataFrame(data)
```

## Generate the actual data

```python
for _ in range(10):
    upload_to_minio(generate_customers(), "raw-customers", f"customers_{str(randint(1000,9999))}.csv")
```

```python
upload_to_minio(generate_products(), "raw-products", "products.csv")
```

```python
for year in range(2015, 2023):
    upload_to_minio(generate_order(), "raw-orders", f"orders_{year}.csv")
```

```python
for year in range(2015, 2023):
    for month in range(1,13,1):
        upload_to_minio(generate_order_detail(), "raw-order-details", f"{year}/orders_{year}-{month:02}.csv")
```

```python
for year in range(2015, 2027):
    for month in range(1,13,1):
        upload_to_minio(generate_logs(), "raw-logs", f"{year}/machine_telemetry_{year}-{month:02}.csv")
```
